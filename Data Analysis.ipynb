### Married customers

```python
married = df[df['Marital Status']=='Married']
married.sort_values(by='Income', ascending=False, axis=0, inplace=True)
married
```

```python
married.shape
```

#### Married customers by income with medsize vehicle size with total claim amount is greater than &200 with less than two policies.

```python
married_people = married[(married['Vehicle Size']=='Medsize') & (married['Total Claim Amount'] > 200) & (married['Number of Policies'] < 2)]
```

```python
married_people
```

#### Indexing Married people whose Marital Status is Married

```python
married_people.set_index('Marital Status', inplace=True)
married_people
```

## Single customers

```python
Single = df[df['Marital Status']=='Single']
Single.sort_values(by='Income', ascending=False, axis=0, inplace=True)
Single
```

```python
Single.shape
```

#### Single customers by income with medsize vehicle size with total claim amount is greater than &200 with less than two policies.

```python
Single_people = Single[(Single['Vehicle Size']=='Medsize') & (Single['Total Claim Amount'] > 200) & (Single['Number of Policies'] < 2)]
```

```python
Single_people
```

### Indexing Single people whose Marital Status is Married
