```python
def namechng(val):
    for i in val:
        if i == 'No':
            val.replace("No", 0, inplace = True)
        elif i == 'Steady':
            val.replace("Steady", 1, inplace = True)
        elif i == 'Up':
            val.replace("Up", 2, inplace = True)
        else:
            val.replace("Down", 3, inplace = True)
```

```python
def convert_to_int(word):
    word_dict = { 'No':0, 'Ch':1}
    return word_dict[word]

train1['tel_48'] = train1['tel_48'].apply(lambda x : convert_to_int(x))
```

```python
def convert_to_int(word):
    word_dict = {'NO':0, '>30':1, '<30': 2}
    return word_dict[word]

train1['tel_49'] = train1['tel_49'].apply(lambda x : convert_to_int(x))
```

```python
# we will use BinaryEncoder from category_encoders library to do binary encoding
import category_encoders as ce
encoder = ce.BinaryEncoder(cols = ['race', 'tel_1', 'tel_2'])
df_binary = encoder.fit_transform(train1)
df_binary.head()
```

```python
# drop rows where values are objects
df_binary[((df_binary.tel_10 != 'V45') &( df_binary.tel_11 != 'V45'))]
```

```python
df_binary = df_binary[pd.to_numeric(df_binary['tel_9'], errors='coerce').notnull()]
df_binary['tel_9'] = df_binary['tel_9'].astype('float')
df_binary['tel_9'] = df_binary['tel_9'].astype('int')
df_binary = df_binary[pd.to_numeric(df_binary['tel_10'], errors='coerce').notnull()]
df_binary['tel_10'] = df_binary['tel_10'].astype('float')
df_binary['tel_10'] = df_binary['tel_10'].astype('int')
df_binary = df_binary[pd.to_numeric(df_binary['tel_11'], errors='coerce').notnull()]
df_binary['tel_11'] = df_binary['tel_11'].astype('float')
df_binary['tel_11'] = df_binary['tel_11'].astype('int')
```

```python
X = df_binary.iloc[:, :-1].values
y = df_binary.iloc[:, -1].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
import sklearn.ensemble as ek
from sklearn import tree
from sklearn import linear_model
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
```

```python
#Evaluating Performance on different types of models
model = { "RandomForest":ek.RandomForestClassifier(n_estimators=50),
        "DecisionTree":tree.DecisionTreeClassifier(max_depth=10),    
         "GradientBoosting":ek.GradientBoostingClassifier(n_estimators=50),
         "GNB":GaussianNB(),
         "LinearRegression":LinearRegression()   
}
```

```python
results = {}
for algo in model:
    clf = model[algo]
    clf.fit(X_train,y_train)
    score = clf.score(X_test,y_test)
    print ("%s : %s " %(algo, score))
    results[algo] = score
```
