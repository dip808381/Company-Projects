```python
def cleaning(val):
    import pandas as pd
    import numpy as np
    from sklearn.model_selection import train_test_split
    from sklearn import preprocessing
    train = pd.read_csv(val)
    train.replace("?", np.nan, inplace = True)
    from sklearn.preprocessing import LabelEncoder #import the LabelEncoder from sklrean library
    le= LabelEncoder() #create the instance of LabelEncoder
    train['gender'] = le.fit_transform(train['gender']) #apply LabelEncoding of country column
    # replace null values by the most frequent values
    train["race"].replace(np.nan, train['race'].value_counts(). index[0], inplace=True)
    train['age'] = train['age'].apply(lambda i: (i[1:3]))
    train = train[pd.to_numeric(train['age'], errors='coerce').notnull()]
    train['age'] = train['age'].astype('int')
    train['age'] = train['age'].apply(lambda i: i+5)
    train1 = train.drop(['patient_id','weight','tel_13', 'tel_14'], axis = 1)
    train1["tel_1"].replace(np.nan, train1['tel_1'].value_counts(). index[0], inplace=True)
    train1["tel_2"] = train1["tel_2"].fillna(method='ffill')
    def namechng(val):
        for i in val:
            if i == 'No':
                val.replace("No", 0, inplace = True)
            elif i == 'Steady':
                val.replace("Steady", 1, inplace = True)
            elif i == 'Up':
                val.replace("Up", 2, inplace = True)
            else:
                val.replace("Down", 3, inplace = True)
    namechng(train1['tel_15'])
    namechng(train1['tel_16'])
    namechng(train1['tel_17'])
    namechng(train1['tel_18'])
    namechng(train1['tel_19'])
    namechng(train1['tel_20'])
    namechng(train1['tel_21'])
    namechng(train1['tel_22'])
    namechng(train1['tel_23'])
    namechng(train1['tel_24'])
    namechng(train1['tel_25'])
    namechng(train1['tel_26'])
    namechng(train1['tel_27'])
    namechng(train1['tel_28'])
    namechng(train1['tel_29'])
    namechng(train1['tel_30'])
    namechng(train1['tel_41'])
    namechng(train1['tel_42'])
    namechng(train1['tel_43'])
    namechng(train1['tel_44'])
    namechng(train1['tel_45'])
    namechng(train1['tel_46'])
    namechng(train1['tel_47'])
    def convert_to_int(word):
        word_dict = { 'No':0, 'Ch':1}
        return word_dict[word]
    train1['tel_48'] = train1['tel_48'].apply(lambda x : convert_to_int(x))
    def convert_to_int(word):
        word_dict = {'NO':0, '>30':1, '<30': 2}
        return word_dict[word]
    train1['tel_49'] = train1['tel_49'].apply(lambda x : convert_to_int(x))
    # we will use BinaryEncoder from category_encoders library to do binary encoding
    import category_encoders as ce
    encoder = ce.BinaryEncoder(cols = ['race', 'tel_1', 'tel_2'])
    df_binary = encoder.fit_transform(train1)
    # drop rows where values are objects
    df_binary[((df_binary.tel_10 != 'V45') &( df_binary.tel_11 != 'V45'))]
    df_binary = df_binary[pd.to_numeric(df_binary['tel_9'], errors='coerce').notnull()]
    df_binary['tel_9'] = df_binary['tel_9'].astype('float')
    df_binary['tel_9'] = df_binary['tel_9'].astype('int')
    df_binary = df_binary[pd.to_numeric(df_binary['tel_10'], errors='coerce').notnull()]
    df_binary['tel_10'] = df_binary['tel_10'].astype('float')
    df_binary['tel_10'] = df_binary['tel_10'].astype('int')
    df_binary = df_binary[pd.to_numeric(df_binary['tel_11'], errors='coerce').notnull()]
    df_binary['tel_11'] = df_binary['tel_11'].astype('float')
    df_binary['tel_11'] = df_binary['tel_11'].astype('int')
    return df_binary
```

```python
import sklearn.ensemble as ek
from sklearn import tree
from sklearn import linear_model
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
```

```python
X = train.loc[:, 'race_0':'tel_49'].values
y = train.loc[:, 'diabetesMed'].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Evaluating Performance on different types of models
model = { "RandomForest":ek.RandomForestClassifier(n_estimators=50),
        "DecisionTree":tree.DecisionTreeClassifier(max_depth=10),    
         "GradientBoosting":ek.GradientBoostingClassifier(n_estimators=50),
         "GNB":GaussianNB(),
         "LinearRegression":LinearRegression()   
}
```

```python
results = {}
for algo in model:
    clf = model[algo]
    clf.fit(X_train,y_train)
    score = clf.score(X_test,y_test)
    print ("%s : %s " %(algo, score))
    results[algo] = score
```
