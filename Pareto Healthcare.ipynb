### Selecting the Independent values for regression model.

```python
data = df2.iloc[:, [3, 5]]
data.head()
```

```python
dummy = pd.get_dummies(df2["Ref Type"])
dummy.head()
```

```python
result = pd.concat([data,dummy], sort=False, axis =1)
result.head()
```

```python
# Bill value is the dependent variable, we are predcting the total cost by the age, sex, and ref-type variables.
result["Bill value"] = df2["Bill value"]
result.head()
```

```python
# Converting the dataframe to the array
X = result.iloc[:, :-1].values           # independent variables
y = result.iloc[:, 15].values            # dependent variables
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# Import the sklearn libraries
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

###### The coefficient of determination is 0.08781171261060405 means that 8 percent of the variance in Y is predictable from X.

## Building a model (Backword Elimination)
#step-1 select a significance level to stay in the model (e.g Sl=0.05)
#step-2 Fill the full model with all possible predictor
#step-3 Consider the predictor with the hightest p-value, if P > Sl=0.05, go to next step otherwise finish
#step-4 Remove the predictor
#step-5 Fit the model without this variable

```python
# Building a optimal model (For the multiple linear regression model we will add an extra column for the the constant value whitch is 65243*1 vector)
X = np.append(arr = np.ones((65243, 1)).astype(int), values = X, axis = 1)
X[:5]
```

```python
# Using Ordinary Least Square Method we will calculate p-value for all independent variables
import statsmodels.api as sm
def backwardElimination(x, sl):
    numVars = len(x[0])
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y, x).fit()
        maxVar = max(regressor_OLS.pvalues).astype(float)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
    regressor_OLS.summary()
    return x
 
SL = 0.05
X_opt = X
X_Modeled = backwardElimination(X_opt, SL)
```

```python
X_Modeled[:5]
```

### Here age variable does not have inportance to predict the Bill Value

```python
X_opt = X_Modeled
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```
