# Extract specific skills from Otherskills column

```python
import pandas as pd # import library to read data into dataframe
import numpy as np # import numpy library
import matplotlib.pyplot as plt  # Visualizing the Data
```

```python
df = pd.read_excel("Data_Science_2020_v2.xlsx") 
df = df.rename(columns = {"Other skills":"otherskills"})
df.head()
```

```python
df.dropna()
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
text = df.otherskills
text 
```

```python
import re

other_skils = ['Machine Learning','Deep Learning','NLP','Statistical Modeling', 'AWS', 'SQL','NoSQL', 'Excel']
join_other_slill = '|'.join(r"\b{}\b".format(x) for x in other_skils)
```

```python
df['otherskill'] = text.str.findall(join_other_slill, flags=re.I).str.join('|')

df.head()
```

```python
df1 = df.drop(['Application_ID','otherskills', 'Institute', 'Degree', 'Current Year Of Graduation', 'Performance_PG', 'Performance_UG', 'Performance_12', 'Performance_10'], axis = 1)
```

```python
#Copying the movie dataframe into a new one since we won't need to use the genre information in our first case.
moviesWithGenres_df = df1.copy()
moviesWithGenres_df.dropna(subset = ["otherskill"], inplace = True)
moviesWithGenres_df.dropna(axis = 0, how='any', thresh=None, subset=["otherskill"], inplace=False)
moviesWithGenres_df.head(50)
```

```python
X  = moviesWithGenres_df[(moviesWithGenres_df['Python (out of 3)'] == 0) & (moviesWithGenres_df['R Programming (out of 3)'] == 0) & (moviesWithGenres_df['Data Science (out of 3)']==0)].index
moviesWithGenres_df.drop(X, inplace = True)
moviesWithGenres_df.dropna()
moviesWithGenres_df.head()
```

```python
moviesWithGenres_df['otherskill'] = moviesWithGenres_df['otherskill'].astype(str)
```

```python
moviesWithGenres_df.info()
```

```python
#Every genre is separated by a | so we simply have to call the split function on |
moviesWithGenres_df['otherskill'] = moviesWithGenres_df.otherskill.str.split('|')
moviesWithGenres_df.head()
```

```python
mdata = moviesWithGenres_df.dropna()
```

```python
mdata.info()
```

```python
#Copying the movie dataframe into a new one since we won't need to use the genre information in our first case.
moviesWithGenres_df = mdata.copy()

#For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
for index, row in mdata.iterrows():
    for genre in row['otherskill']:
        moviesWithGenres_df.at[index, genre] = 1
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
pd.set_option('display.max_columns', 100)
moviesWithGenres_df.head(40)
```

```python
new = moviesWithGenres_df.drop(['Current City','Stream', 'otherskill'], axis = 1)
new.head()
```

```python
X1 = new.iloc[:, [0,1,2,3,4,5,7,8,9,10,11]].values  # Values converted to array
X1
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X1)
```

```python
k_means_labels1 = kmeans.labels_
k_means_labels1
```

```python
# Adding labelis in the given dataset
new1 = new.iloc[:, 0:11]
new1["Data Sientist"] = k_means_labels1
new1.head(5)
```
